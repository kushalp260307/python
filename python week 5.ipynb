```python
my_list = [10, 3.5, "hello", True]

print("List:", my_list)
print("Type of list:", type(my_list))

for item in my_list:
    print("Element:", item, "Type:", type(item))
```

```python
a = [1, 2, 3]
b = [4, 5, 6]

# Concatenation
print("Concatenation:", a + b)

# Repetition
print("Repetition:", a * 3)

# Membership
print("Is 2 in list a?", 2 in a)

# Length
print("Length of list a:", len(a))
```

```python
nums = [10, 5, 20, 8, 15]

print("Maximum:", max(nums))
print("Minimum:", min(nums))
print("Sum:", sum(nums))

# Sorting
print("Ascending:", sorted(nums))
print("Descending:", sorted(nums, reverse=True))

# Index of element 8
print("Index of 8:", nums.index(8))
```

```python
lst = [10, 20, 30]

# Append
lst.append(40)
print("After append:", lst)

# Insert
lst.insert(1, 15)
print("After insert:", lst)

# Remove
lst.remove(20)
print("After remove:", lst)

# Pop
lst.pop()
print("After pop:", lst)

# Reverse
lst.reverse()
print("After reverse:", lst)
```

```python
# Taking input
nums = list(map(int, input("Enter numbers separated by space: ").split()))

# Even numbers
even = [x for x in nums if x % 2 == 0]
print("Even numbers:", even)

# Odd numbers
odd = [x for x in nums if x % 2 != 0]
print("Odd numbers:", odd)

# Largest and smallest
print("Largest:", max(nums))
print("Smallest:", min(nums))

# Bonus: Remove duplicates
unique = list(set(nums))
print("Without duplicates:", unique)
```
