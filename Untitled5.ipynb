```python
# Library Management System using OOP Concepts

class Book:
    def __init__(self, title, author, book_id):
        self.title = title
        self.author = author
        self.book_id = book_id

    def display_book(self):
        print("Book ID :", self.book_id)
        print("Title   :", self.title)
        print("Author  :", self.author)
        print("-" * 30)


class Library:
    def __init__(self):
        self.books = []

    # Add a book
    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added successfully!\n")

    # Display all books
    def display_books(self):
        if len(self.books) == 0:
            print("No books available in the library.")
        else:
            print("\nLibrary Book Details")
            print("=" * 30)
            for book in self.books:
                book.display_book()


# Main Program
library = Library()

# Creating book objects
b1 = Book("Python Programming", "John Smith", 101)
b2 = Book("Data Structures", "David Miller", 102)
b3 = Book("AI Basics", "Andrew Lee", 103)

# Adding books to library
library.add_book(b1)
library.add_book(b2)
library.add_book(b3)

# Displaying all books
library.display_books()
```

```python
# Bank Management System using OOP Concepts

class BankAccount:
    def __init__(self, name, account_number, balance=0):
        self.name = name
        self.account_number = account_number
        self.balance = balance

    # Deposit money
    def deposit(self, amount):
        self.balance += amount
        print(f"₹{amount} deposited successfully.")

    # Withdraw money
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance!")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")

    # Display account details
    def display(self):
        print("\nAccount Details")
        print("-" * 30)
        print("Account Holder :", self.name)
        print("Account Number :", self.account_number)
        print("Balance        : ₹", self.balance)


# Main Program
acc1 = BankAccount("Kushal", 123456789, 5000)

while True:
    print("\n===== Bank Management System =====")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Display Account Details")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        amount = float(input("Enter amount to deposit: "))
        acc1.deposit(amount)

    elif choice == 2:
        amount = float(input("Enter amount to withdraw: "))
        acc1.withdraw(amount)

    elif choice == 3:
        acc1.display()

    elif choice == 4:
        print("Thank you for using Bank Management System!")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
# Student Result Management System using OOP Concepts

class Student:
    def __init__(self, name, usn, marks1, marks2, marks3):
        self.name = name
        self.usn = usn
        self.marks1 = marks1
        self.marks2 = marks2
        self.marks3 = marks3

    # Calculate total marks
    def total_marks(self):
        return self.marks1 + self.marks2 + self.marks3

    # Calculate percentage
    def percentage(self):
        return self.total_marks() / 3

    # Calculate grade
    def grade(self):
        per = self.percentage()

        if per >= 90:
            return "A+"
        elif per >= 75:
            return "A"
        elif per >= 60:
            return "B"
        elif per >= 40:
            return "C"
        else:
            return "Fail"

    # Display result
    def display_result(self):
        print("\n===== Student Result =====")
        print("Student Name :", self.name)
        print("USN          :", self.usn)
        print("Marks 1      :", self.marks1)
        print("Marks 2      :", self.marks2)
        print("Marks 3      :", self.marks3)
        print("Total Marks  :", self.total_marks())
        print("Percentage   :", self.percentage(), "%")
        print("Grade        :", self.grade())


# Main Program
name = input("Enter student name: ")
usn = input("Enter USN: ")

m1 = int(input("Enter marks in Subject 1: "))
m2 = int(input("Enter marks in Subject 2: "))
m3 = int(input("Enter marks in Subject 3: "))

# Creating object
s1 = Student(name, usn, m1, m2, m3)

# Displaying result
s1.display_result()
```
